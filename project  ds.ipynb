```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv("D:\StudentsPerformance.csv")
data.head(10)
```

```python
data["ID"] = np.arange(1, len(data["ID"])+1 )
data.head(10)
```

```python
def rates(subject):
    rank = []
    for (i) in data[f"{subject} score"]:
        if i > 85:
            rank.append("A")
        elif 75 < i <= 85:
            rank.append("B")
        elif 65 < i <= 75:
            rank.append("C")
        elif 50 < i <= 65:
            rank.append("D")
        else:
            rank.append("F")
            
    data[f"{subject} rate"] = rank
    
subjects = ['programming', 'physics', 'algebra']
for subject in subjects:
    rates(subject)

data.head(10)
```

```python
new_columns=["Student Name","gender","ID","algebra score","algebra rate","programming score","programming rate","physics score","physics rate"]
data=data.reindex(columns=new_columns)
data.head(10)
```

```python
data.sort_values("algebra score", ascending = False ).head(10)
# data.sort_values("pyhsics score", ascending = False ).head(10)
# data.sort_values("programming score", ascending = False ).head(10)
```

```python
def Top_Student(subject, data):
    print(f"Top students in {subject} are: ")
    return data.sort_values(f"{subject} score", ascending=False).head(5) 
     
Top_Student("algebra",data)
# Top_Student("physics",data)
# Top_Student("programming",data)
```

```python
data["total grades"] = data[["algebra score", "programming score", "physics score"]].sum(axis=1)
data
```

```python
id = data['ID'].head(5)
grades = data['total grades'].head(5)

plt.style.use("classic")

plt.bar(id.values, grades.values, color="k",width=0.5, label="Total Grades")

plt.xlabel('ID')
plt.ylabel('students total grades')
plt.legend()
plt.grid(True)
plt.show()
print(id.values)
```

```python
Rates=["A","B","C","D"]
arr1=[] 
arr2=[] 
arr3=[] 
for i in Rates:
    arr1.append(np.sum(data["algebra rate"].values == i))
    arr2.append(np.sum(data["physics rate"].values == i))
    arr3.append(np.sum(data["programming rate"].values == i))
    
# grades total for A B C D in each subject :
print(arr1)
print(arr2)
print(arr3)
```
